# Text extraction

```python
import os
import PyPDF2
import re
pdfFiles = []
for filename in os.listdir('./kanerai_docs_for_columbia_confidential'):
    if filename.endswith('.pdf'):
        pdfFiles.append('./kanerai_docs_for_columbia_confidential/' + filename)
success = 0
all_text = ''
for i in pdfFiles:
    try:
        pdfFileObj = open(i,'rb') 
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        text = ''
        for j in range(pdfReader.getNumPages() - (7+5)):
            text = text + str(pdfReader.getPage(j + 7).extractText())
        print(i)
        if len(text) > 200000:
            print(str(len(text)) + ' characters')
            success += 1
        else:
            print('failed to read properly')
        all_text = all_text + text
    except:
        pass
success / len(pdfFiles)
```

# Word2vec model

```python
import nltk
import urllib
import re
from gensim.models import Word2Vec

text = all_text

# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
text = re.sub(r'section','',text)
text = re.sub(r'\([^)]*\)','',text)

sentences = nltk.sent_tokenize(text)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

for i in sentences[:]:
    if len(i) == 1:
        sentences.remove(i)
    else:
        j.replace()
        for j in i[:]:
            if j == '.':
                i.remove('.')
            elif j == '..':
                i.remove('..')
            elif j == '...':
                i.remove('...')
            elif j=='_':
                i.remove('_')
            elif j=='-':
                i.remove('-')
```

```python
# train different word2vec models
for size in [25, 64, 100, 200]:
    model = Word2Vec(sentences, size=size, alpha=0.025, window=5, sg=1, min_count=1)
    filename = 'w2vmodels/w2v_%d_%s.bin' %(size,'sg')
    model.save(filename)
```
